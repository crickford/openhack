# Loop through each image for resizing and padding

```python
import os
import shutil
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Needed to display matplotlib plots in Jupyter
%matplotlib inline

# Create a figure to display the images
fig = plt.figure(figsize=(12, 16))

# loop recursively through the folders
dir_num = 0
imgdir = 'gear_images'
resizedDir = '/data/home/team13/notebooks/resized_images/'
desired_size  = 128

for subdir, dirs, files in os.walk(imgdir):
    for file in files:
        filepath = subdir + os.sep + file
        folder = subdir.rsplit('/', 1)
        # Create path for Resized File
        resizedFileNameAndPath = os.path.join(resizedDir, folder[-1], file)
        img = Image.open(filepath)
        old_size = img.size  # old_size[0] is in (width, height) format
        ratio = float(desired_size)/max(old_size)
        new_size = tuple([int(x*ratio) for x in old_size])
        imgThumb = img.resize(new_size, Image.ANTIALIAS)

        new_im = Image.new("RGB", (desired_size, desired_size), "#fff")
        new_im.paste(imgThumb, ((desired_size-new_size[0])//2,
                   (desired_size-new_size[1])//2))
        
        new_im.save(resizedFileNameAndPath)
```
